## 2. 2. Fill missing values in the Embarked column with the most frequent value (mode).Dropping Rows and Columns:

```python
# Fill missing values in the 'Embarked' column with the most frequent value (mode)
most_frequent_value = df['Embarked'].mode()[0]
df['Embarked'].fillna(most_frequent_value, inplace=True)
```

```python
# Dropping rows with any missing values
df_dropped_rows = df.dropna()
```

```python
# Dropping columns with any missing values
df_dropped_columns = df.dropna(axis=1)
```

## 3. Drop the Cabin column due to a high number of missing values.

```python
df.drop(columns=['Cabin'], inplace=True)
```

## 4. Drop rows where Fare is missing.

```python
df.dropna(subset=['Fare'], inplace=True)
```

## 5. Convert the Age column to an integer type.


```python
df['Age'] = df['Age'].fillna(df['Age'].median()).astype('int')
```

## 6.  Convert the number data in the Survived column in the data to categorical data (Hint: Assign dead to 0 ,alive to 1).

```python
# Convert the 'Survived' column to categorical data (0 -> 'dead', 1 -> 'alive')
df['Survived'] = df['Survived'].map({0: 'dead', 1: 'alive'})
```

## 1. How Many People Survived And How Many People died?

```python
survival_counts = df['Survived'].value_counts()
print("Number of people who are dead:", survival_counts['dead'])
print("Number of people who are alive:", survival_counts['alive'])
```

```python
# Count the number of passengers in each class
class_counts = df['Pclass'].value_counts()

# Display the counts
for pclass, count in class_counts.items():
    print(f"Number of passengers in class {pclass}: {count}")
```
